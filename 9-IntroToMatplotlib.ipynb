```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [18, 19, 20, 20, 20, 23, 24, 25]
```

```python
weight_list = [150, 160, 160, 160, 170, 175, 180, 185]
```

```python
plt.plot(age_list, weight_list,"y")
# Görseli kaydetmek için plt.savefig("gorsel.png")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
np_age_list = np.array(age_list)
```

```python
np_age_list
```

```python
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list, np_weight_list,"g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0, 10, 20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2,"m*")
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2,"m-")
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2,"m*-")
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2,"m--")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(numpy_arr1, numpy_arr2,"m*-")
plt.subplot(1, 2, 2)
plt.plot(numpy_arr2, numpy_arr1,"m--")
plt.show()
```

```python
my_figure = plt.figure()
figure_ax = my_figure.add_axes([0.1,0.1,0.3,0.3])
figure_ax.plot(numpy_arr1, numpy_arr2,"m*-")
figure_ax.set_xlabel("X Label")
figure_ax.set_ylabel("Y Label")
figure_ax.set_title("Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=100)
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_arr1, numpy_arr1 ** 2,label = "numpy array **2")
new_axes.plot(numpy_arr1, numpy_arr1 ** 3,label = "numpy array **3")
new_axes.legend(loc=0)
plt.show()
```
