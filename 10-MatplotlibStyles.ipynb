```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = np.linspace(0, 10, 20)
```

```python
data1
```

```python
data2 = data1 ** 2
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2, color = "#1E0708", alpha = 0.5)
# alpha transparanlık değeri verir 0 ile 1 arasında
my_axes.plot(data2, data1, color = "#53CA0A")
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()

new_axes.plot(data1, data1 + 2, color = "blue", linewidth = 1.0)
new_axes.plot(data1, data1 + 3, color = "yellow", linewidth = 4.0)

new_axes.plot(data1, data1 + 4, color = "#C85D09", linestyle = "-.")
new_axes.plot(data1, data1 + 5, color = "#C85D09", linestyle = ":")
new_axes.plot(data1, data1 + 6, color = "#C85D09", linestyle = "--")

new_axes.plot(data1, data1 + 7, color = "#000000", linestyle = "--", marker = "o", markersize = 10, markerfacecolor = "red", markeredgewidth = 3, markeredgecolor = "blue")
new_axes.plot(data1, data1 + 8, color = "#000000", linestyle = "--", marker = "+", markersize = 4)

plt.show()
```

```python
plt.scatter(data1, data2)
plt.show()
```
