```python
# Arithmetic
contest_result = pd.Series(data=[10, 5, 100], index=["Ali", "Ahmet", "Ayşe"])
contest_result2 = pd.Series(data=[20, 50, 10], index=["Ali", "Ahmet", "Ayşe"])
```

```python
contest_result
```

```python
contest_result["Ali"]
```
