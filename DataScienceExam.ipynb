### Soru 2
Aşağıdaki iki NumPy dizisi verilmiştir:

```python
a = np.array([10, 20, 30, 40, 50])
b = np.array([1, 2, 3, 4, 5])
```

Bu iki dizi üzerinde **eleman bazında toplama, çıkarma, çarpma ve bölme** işlemlerini yapın.
Örneğin 10 + 1 = 11 şeklinde index'i denk gelen tüm elemanlar birbiriyle işleme girip yeni dizide yerini almalıdır.
Her sonucu ayrı satırda yazdırın.

````python
# Cevabınızı buraya yazın
a = np.array([10, 20, 30, 40, 50])
b = np.array([1, 2, 3, 4, 5])
toplam_dizisi = a + b
cikarma_dizisi = a - b
carpma_dizisi = a * b
bolme_dizisi = a / b
print("Toplam Dizisi:", toplam_dizisi)
print("Çıkarma Dizisi:", cikarma_dizisi)
print("Çarpma Dizisi:", carpma_dizisi)
print("Bölme Dizisi:", bolme_dizisi)
````

### Soru 3
0 ile 100 arasında **rastgele 15 tam sayıdan** oluşan bir NumPy dizisi oluşturun (`np.random.randint` kullanın, seed=42 ayarlayın).  
Bu dizinin **ortalama (mean)**, **standart sapma (std)**, **minimum** ve **maksimum** değerlerini hesaplayıp yazdırın.

````python
# Cevabınızı buraya yazın
np.random.seed(42)
array = np.random.randint(0,100, size=15)
ortalama = np.mean(array)
standart_sapma = np.std(array)
minimum = np.min(array)
maksimum = np.max(array)
print("Dizi:", array)
print("Ortalama:", ortalama)
print("Standart Sapma:", standart_sapma)
print("Minimum:", minimum)
print("Maksimum:", maksimum)
````

### Soru 7
Aşağıdaki iki DataFrame'i **"id"** sütununa göre birleştirin (merge) ve sonucu yazdırın:

```python
df1 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "ürün": ["Laptop", "Telefon", "Tablet", "Kulaklık"]
})

df2 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "fiyat": [15000, 10000, 5000, 500],
    "stok": [30, 50, 25, 100]
})
```

Birleştirme sonrasında **fiyatı 5000'den büyük** olan ürünleri filtreleyin.

````python
# Cevabınızı buraya yazın
df1 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "ürün": ["Laptop", "Telefon", "Tablet", "Kulaklık"]
})

df2 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "fiyat": [15000, 10000, 5000, 500],
    "stok": [30, 50, 25, 100]
})

birlesik_df = pd.merge(df1, df2, on="id")
birlesik_df

filtrelenmis_df = birlesik_df[birlesik_df["fiyat"] > 5000]
filtrelenmis_df
````

## Bölüm 3: Matplotlib

### Soru 8
Aşağıdaki verileri kullanarak bir **çizgi grafiği (line plot)** çizin:

```python
aylar = ["Oca", "Şub", "Mar", "Nis", "May", "Haz"]
satislar = [120, 135, 148, 160, 175, 190]
```

Grafik gereksinimleri:
- Başlık: "Aylık Satış Trendi"
- X ekseni etiketi: "Aylar"
- Y ekseni etiketi: "Satış Adedi"
- Çizgi rengi kırmızı, işaretçi (marker) yuvarlak olsun

````python
# Cevabınızı buraya yazın
aylar = ["Oca", "Şub", "Mar", "Nis", "May", "Haz"]
satislar = [120, 135, 148, 160, 175, 190]

plt.plot(aylar, satislar, color='red', marker='o')
plt.title("Aylık Satış Trendi")
plt.xlabel("Aylar")
plt.ylabel("Satış Adedi")
plt.grid(True)
plt.show()
````

### Soru 9
Aşağıdaki verileri kullanarak bir **bar (çubuk) grafiği** çizin:

```python
diller = ["Python", "JavaScript", "Java", "C++", "Go"]
kullanici = [35, 25, 20, 12, 8]
```

Grafik gereksinimleri:
- Başlık: "Programlama Dilleri Popülerliği"
- Her çubuk farklı renkte olsun
- Figür boyutu 10×6 olsun

````python
# Cevabınızı buraya yazın
diller = ["Python", "JavaScript", "Java", "C++", "Go"]
kullanici = [35, 25, 20, 12, 8]

plt.figure(figsize=(10, 6))
renkler = ['blue', 'orange', 'green', 'red', 'purple']
plt.bar(diller, kullanici, color=renkler)
plt.title("Programlama Dilleri Popülerliği")
plt.xlabel("Programlama Dilleri")
plt.ylabel("Kullanıcı")
plt.show()
````

## Bölüm 4: Seaborn

### Soru 10
Seaborn'un yerleşik **"tips"** veri setini yükleyin (`sns.load_dataset("tips")`).

1. İlk 5 satırı gösterin.
2. `total_bill` sütunu için bir **histogram (histplot)** çizin.
3. Grafik başlığını "Toplam Hesap Dağılımı" olarak ayarlayın ve KDE eğrisini ekleyin.

````python
# Cevabınızı buraya yazın
tips = sns.load_dataset("tips")
print(tips.head())

plt.figure(figsize=(8, 5)) 
sns.histplot(data=tips, x="total_bill", kde=True, color="teal")
plt.title("Toplam Hesap Dağılımı")
plt.xlabel("Toplam Hesap ($)")
plt.ylabel("Frekans")
plt.show()
````

### Soru 11
Tips veri setini kullanarak **`total_bill`** ve **`tip`** arasındaki ilişkiyi gösteren bir **scatter plot (saçılım grafiği)** çizin (`sns.scatterplot`).

Grafik gereksinimleri:
- Noktaları **`smoker`** (sigara içen/içmeyen) kategorisine göre renklendirin (hue).
- Başlık: "Hesap vs Bahşiş İlişkisi"
- Figür boyutu 10×6 olsun

````python
# Cevabınızı buraya yazın
tips = sns.load_dataset("tips")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker")
plt.title("Hesap vs Bahşiş İlişkisi")
plt.xlabel("Toplam Hesap")
plt.ylabel("Bahşiş")
plt.show()
````
