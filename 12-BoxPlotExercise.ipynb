```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])
plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
plt.figure(figsize=(13,8))
plt.subplot(1,2,1)
sns.boxplot(x="class", y="age", data=df)
plt.title("Box Plot of Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")
plt.subplot(1,2,2)
sns.boxplot(x="class", y="fare", data=df)
plt.title("Box Plot of Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```
